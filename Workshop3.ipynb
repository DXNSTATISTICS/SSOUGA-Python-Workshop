# Bar Charts

```python
Names = ["Tom Brady", "Aaron Rodgers", "Matt Ryan", "Bryan Dixon", "Russell Wilson"]
SuperBowlWins = [7,1,0,0,1]
SuperBowlAppearances = [10,1,1,0,2]
```

```python
plt.bar(Names,SuperBowlWins)
plt.title("Super Bowl Wins")
plt.ylabel("Wins")
plt.xticks(rotation=45)
plt.show()

#plt.savefig('superbowlwins.png')
```

```python
plt.scatter(SuperBowlWins,SuperBowlAppearances)
```

```python
fig, ax = plt.subplots()
ax.scatter(SuperBowlWins, SuperBowlAppearances)
ax.set_title("Wins Vs. Appearances")
ax.set_xlabel("Wins")
ax.set_ylabel("Appearances")

for i, txt in enumerate(Names):
    ax.annotate(txt, (SuperBowlWins[i], SuperBowlAppearances[i]))
```

# fig, ax = plt.subplots()

plt.subplots() is a function that returns a tuple containing a figure and axes object(s). Thus when using fig, ax = plt.subplots() you unpack this tuple into the variables fig and ax. Having fig is useful if you want to change figure-level attributes or save the figure as an image file later (e.g. with fig.savefig('yourfilename.png')). You certainly don't have to use the returned figure object but many people do use it later so it's common to see. 

```python
import numpy as np

Names = ["Tom Brady", "Aaron Rodgers", "Matt Ryan", "Bryan Dixon", "Russell Wilson"]
SuperBowlWins = [7,1,0,0,1]
SuperBowlAppearances = [10,1,1,0,2]

X_axis = np.arange(len(Names))
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(X_axis - width/2, SuperBowlWins, width, label='SB Wins')
rects2 = ax.bar(X_axis + width/2, SuperBowlAppearances, width, label='SB Appearances')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Count')
ax.set_title('Wins VS Appearances')
ax.set_xticks(X_axis,Names,rotation = 45)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()
plt.rcParams["figure.figsize"] = (5,5)

plt.show()
```
